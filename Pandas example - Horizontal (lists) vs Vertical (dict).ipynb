```python
import pandas as pd
```

### Vertical Data (Multiple seperate lists)

```python
# This is what you have
authors = [1,1,1]
book    = [2,2,2]
cost    = [3,3,3]
```

```python
# create a dictionary
{'author':authors,
 'book':book,
 'cost':cost}
```

```python
# Vertical data (Basically yoou just put dataframe around a dictionary)
df = pd.DataFrame({'author':authors,
                   'book':book,
                   'cost':cost})
df
```

```python
# This is what you are doing right now
results = []
author_list = []
book_list = []
cost_list = []
for i in range(10):
    author_list.append(1)  ## 1,2,3 would be different every time, just same for example. These woudl be cost etc.
    book_list.append(2) 
    cost_list.append(3) 

# Turn into dataframe
df = pd.DataFrame({'author':author_list,
                   'book':book_list,
                   'cost':cost_list})

df
```

### Horizontal Data (list of lists)

```python
a = [1,2,3]
b = [1,2,3]
c = [1,2,3]
```

```python
# Horisotal data - turn this into a data frame
df2 = pd.DataFrame([a,b,c], columns = ['author','book','cost'])
df2
```

```python
## Iterating over on a Row by Row basis - this is what you _could_ do
results = []
for i in range(10):
    results.append([1,2,3])  ## 1,2,3 would be different every time, just same for example. These woudl be cost etc.

# Turn into dataframe
df2 = pd.DataFrame(results, columns = ['author','book','cost'])
df2
```
